# embedding live video player with play and pause button

```python
import cv2
import numpy as np
import tkinter as tk

# Create a function to read and display live video
def show_video():
    # Create a video capture object
    cap = cv2.VideoCapture(0)

    # Check if the camera is opened successfully
    if not cap.isOpened():
        print("Error opening video stream or file")

    # Create a window for displaying the video
    cv2.namedWindow("Live Video", cv2.WINDOW_NORMAL)

    # Loop through the video frames
    while True:
        # Read the frame from the video
        ret, frame = cap.read()

        # Check if the frame was successfully read
        if not ret:
            print("Error reading frame")
            break

        # Display the frame in the window
        cv2.imshow("Live Video", frame)

        # Wait for a key press
        key = cv2.waitKey(1)

        # Check if the 'q' key was pressed to quit
        if key == ord('q'):
            break

        # Check if the 'p' key was pressed to pause
        if key == ord('p'):
            # Pause the video
            while True:
                key = cv2.waitKey(1)

                # Check if the 'r' key was pressed to resume
                if key == ord('r'):
                    break

    # Release the video capture object and close the window
    cap.release()
    cv2.destroyAllWindows()

# Create a function to pause the video
def pause_video():
    cv2.waitKey(0)

# Create a function to resume the video
def resume_video():
    pass

# Create a GUI window
root = tk.Tk()

# Set the window title and size
root.title("Video Player")
root.geometry("300x100")

# Create a button to pause the video
pause_button = tk.Button(root, text="Pause", command=pause_video)
pause_button.pack(side="left", padx=10, pady=10)

# Create a button to resume the video
resume_button = tk.Button(root, text="Resume", command=resume_video)
resume_button.pack(side="right", padx=10, pady=10)

# Start the video display function
show_video()

# Start the GUI event loop
root.mainloop()
```

```python
import cv2
import tkinter as tk
from PIL import Image, ImageTk

def play_live_video():
    cap = cv2.VideoCapture(0)
    root = tk.Tk()
    canvas = tk.Canvas(root, width=cap.get(cv2.CAP_PROP_FRAME_WIDTH),
                       height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    canvas.pack()
    play_button = tk.Button(root, text="Play", command=lambda: play(cap, canvas))
    play_button.pack(side=tk.LEFT)
    pause_button = tk.Button(root, text="Pause", command=lambda: pause(cap))
    pause_button.pack(side=tk.LEFT)

    def play(cap, canvas):
        ret, frame = cap.read()
        if ret:
            photo = ImageTk.PhotoImage(image=Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)))
            canvas.image = photo  # keep a reference to the photo to prevent garbage collection
            canvas.create_image(0, 0, image=photo, anchor=tk.NW)
            root.after(10, lambda: play(cap, canvas))

    def pause(cap):
        cap.release()

    root.mainloop()

if __name__ == "__main__":
    play_live_video()
```

```python
import cv2
import tkinter as tk
from PIL import Image, ImageTk

class VideoPlayer:
    def __init__(self):
        self.cap = cv2.VideoCapture(0)
        self.root = tk.Tk()
        self.canvas = tk.Canvas(self.root, width=self.cap.get(cv2.CAP_PROP_FRAME_WIDTH), 
                                height=self.cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        self.canvas.pack()
        self.play_button = tk.Button(self.root, text="Play", command=self.play)
        self.play_button.pack(side=tk.LEFT)
        self.pause_button = tk.Button(self.root, text="Pause", command=self.pause)
        self.pause_button.pack(side=tk.LEFT)

    def play(self):
        ret, frame = self.cap.read()
        if ret:
            self.photo = ImageTk.PhotoImage(image=Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tk.NW)
            self.root.after(10, self.play)

    def pause(self):
        self.cap.release()

    def run(self):
        self.root.mainloop()

if __name__ == "__main__":
    player = VideoPlayer()
    player.run()
```
